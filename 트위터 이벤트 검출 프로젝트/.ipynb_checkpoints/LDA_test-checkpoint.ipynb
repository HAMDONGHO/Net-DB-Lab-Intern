```python
data = pd.read_csv('/Users/hamdongho/WebProject/ITDB/트위터 이벤트 검출 프로젝트/코로나데이터/COVID_data_2020-08-01_to_2020-08-03.csv', error_bad_lines=False)
text_data_pd = data[['text']]

okt = Okt()
```

```python
text_data_pd
```

```python
text_data_pd['text'] = text_data_pd.apply(lambda row: okt.nouns(row['text']), axis=1)

#for i in range(len(text_data_pd)):
#    value = okt.nouns(text_data_pd[i])
#    text_data_pd[i] = value
```

```python
text_data_pd.head(5)
```

```python
tokenized_doc = text_data_pd['text'].apply(lambda x: [word for word in x if len(word) > 1])
```

```python
# 역토큰화 (토큰화 작업을 되돌림)
detokenized_doc = []
for i in range(len(text_data_pd)):
    t = ' '.join(tokenized_doc[i])
    detokenized_doc.append(t)

text_data_pd['text'] = detokenized_doc # 다시 text['headline_text']에 재저장
```

```python
text_data_pd
```

```python
X = TfidfVectorizer().fit_transform(text_data_pd['text'])
X.shape # TF-IDF 행렬의 크기 확인
```

```python
from sklearn.decomposition import LatentDirichletAllocation
lda_model=LatentDirichletAllocation(n_components=10,learning_method='online',random_state=777,max_iter=1)
```

```python
lda_top=lda_model.fit_transform(X)
print(lda_model.components_)
print(lda_model.components_.shape) 
```

```python
terms = TfidfVectorizer().get_feature_names() # 단어 집합. 1,000개의 단어가 저장됨.

def get_topics(components, feature_names, n=5):
    for idx, topic in enumerate(components):
        print("Topic %d:" % (idx+1), [(feature_names[i], topic[i].round(2)) for i in topic.argsort()[:-n - 1:-1]])
get_topics(lda_model.components_,terms)
```
